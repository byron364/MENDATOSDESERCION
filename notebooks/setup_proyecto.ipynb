```python
from pathlib import Path

# Nombre del proyecto
proyecto = "MENDATOSDESERCION"

# Ruta base
ruta_base = Path(proyecto)

print(f"Proyecto: {proyecto}")
```

```python
carpetas = [
    "data/raw",
    "data/processed",
    "notebooks",
    "src",
    "models",
    "reports",
    "reports/figures"
]

for carpeta in carpetas:
    (ruta_base / carpeta).mkdir(parents=True, exist_ok=True)

print(" Estructura creada correctamente")
```

```python
import os

ruta_venv = ruta_base / ".venv"

if not ruta_venv.exists():
    os.system(f"python -m venv {ruta_venv}")
    print("Entorno virtual creado")
else:
    print("El entorno virtual ya existe")
```

```python
for ruta in ruta_base.rglob("*"):
    print(ruta)
```

```python
import platform

sistema = platform.system()

print("Activar el entorno virtual desde la terminal:\n")

if sistema == "Windows":
    print(fr"{proyecto}\.venv\Scripts\activate")
else:
    print(f"source {proyecto}/.venv/bin/activate")
```
